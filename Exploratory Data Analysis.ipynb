<h3>Vizualization</h3>

<p><b>Seaborn</b> merupakan salah satu pustaka visual Python yang berlandaskan pada matplotlib</p>
<p><b>Seaborn</b> menyediakan antar-muka tingkat tinggi untuk menangani permasalahan terkait visualisasi data secara statistik agar tampak lebih menarik</p>

```python
#Menampilkan jumlah persentase pada fitur gender antara female dengan male
sns.set(style='whitegrid')
ax=sns.barplot(x=data['gender'].value_counts().index,y=data['gender'].value_counts().values,palette="Blues_d",hue=['female','male'])
plt.legend(loc=8)
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Show of Gender Bar Plot')
plt.show()
```

Visualisasi diatas berdasarkan hasil dari persentase pada fitur gender, mendapatkan insight bahwasahnya
gender atau jenis kelamin lebih dominan bergender female sebanyak 518, sedangkan male 482.

```python
plt.figure(figsize=(7,7))
sns.barplot(x=data['race/ethnicity'].value_counts().index,
              y=data['race/ethnicity'].value_counts().values)
plt.xlabel('Race/Ethnicity')
plt.ylabel('Frequency')
plt.title('Show of Race/Ethnicity Bar Plot')
plt.show()
```

Berdasarkan visualisasi diatas dapat dilihat pula peringkat group yang paling tertinggi yakni <b>group C</b>, sedangkan group yang paling terendah yakni <b>group A<b>.

```python
#Visualisasi hasil dari writing score berdasarkan jenis level jenjang pendidikan
plt.figure(figsize=(10,7))
sns.barplot(x = "parental level of education", y = "writing score", hue = "gender", data = data)
plt.xticks(rotation=45)
plt.show()
```

Visualisasi diatas menggambarkan presentase hasil dari test writing score, berdasarkan dari jenis jenjang pendidikan masing-masing personal 
1. bachelor's degree lebih dominan gender female
2. some college lebih dominan gender female
3. master's degree lebih dominan gender female
4. associate's degree lebih dominan gender female
5. high school lebih dominan gender female
6. some high school lebih dominan gender female

```python
#Visualisasi hasil dari reading score berdasrkan level jenjang pendidikan
plt.figure(figsize=(10,7))
sns.barplot(x = "parental level of education", y = "reading score", hue = "gender", data = data)
plt.xticks(rotation=45)
plt.show()
```

```python
#Visualisasi hasil dari math score berdasarkan level jenjang pendidikan
plt.figure(figsize=(10,7))
sns.barplot(x = "parental level of education", y = "math score", hue = "gender", data = data)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.catplot(x="gender", y="math score",
                 hue="test preparation course",
                 data=data, kind="bar")
plt.title('for Test Preparation Course Gender & Math_Score')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.title("Lunch - Free/reduced & standard")
sns.barplot(x=data lunch score.index,y=data lunch score.values)
plt.show()
```

```python
f,ax=plt.subplots(figsize=(9,10))
sns.barplot(x=data['gender'].value_counts().values,y=data['gender'].value_counts().index,alpha=0.5,color='red',label='Gender')
sns.barplot(x=data['race/ethnicity'].value_counts().values,y=data['race/ethnicity'].value_counts().index,color='blue',alpha=0.7,label='Race/Ethnicity')
ax.legend(loc='upper right',frameon=True)
ax.set(xlabel='gender , race/ethnicity',ylabel='Groups',title="Gender vs Race/Ethnicity ")
plt.show()
```

<h3>Pie Chart</h3>

```python
labels=data['race/ethnicity'].value_counts().index
colors=['silver','blue','yellow','green','orange']
explode=[0,0,0.1,0,0]
values=data['race/ethnicity'].value_counts().values

#visualization
plt.figure(figsize=(7,7))
plt.pie(values,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%')
plt.title('Race/Ethnicity According Analysis',color='black',fontsize=10)
plt.show()
```
